```python
#Dependencies

import matplotlib.pyplot as plt
import pandas as pd
import statistics
import numpy as np
from matplotlib.ticker import MultipleLocator
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

y_axis_l=[]
for x in y_axis:
    y_axis_l.append(x*5)
y_axis_l
```

```python
fig1,PyBerFare=plt.subplots()
PyBerFare.bar(x_axis, y_axis , color="blue", linewidth=2, label='Boston')
PyBerFare.set_xlabel("Date")
PyBerFare.set_ylabel("Fare in $")
PyBerFare.set_xlim(-0.75,11.75)
PyBerFare.set_ylim(0,50)
#PyBerFare.grid()
PyBerFare.legend()
```

```python
plt.scatter(x_axis,y_axis,s=y_axis_l,color="red")
```

```python
fig1,PyBerFareScatter=plt.subplots()
PyBerFareScatter.scatter(y_axis,x_axis,s=y_axis_l,linewidth=2,c="skyblue",alpha=0.8,edgecolor="black")
PyBerFareScatter.legend("Boston")
fig1.gca().invert_yaxis()
PyBerFareScatter.set_xlim(0,60,10)
PyBerFareScatter.set_title("Pyber Fare By Month")
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig,PyBerPie=plt.subplots()
PyBerPie.pie(y_axis,labels = x_axis,explode = explode_values,autopct='%.1f%%')
#PyBerPie.set_figsize(8, 8)

plt.show()


```

```python
stdev=statistics.stdev(y_axis)
```
